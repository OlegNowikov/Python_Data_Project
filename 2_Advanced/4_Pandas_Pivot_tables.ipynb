```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df_original = dataset['train'].to_pandas()
# Data Cleansing
df_original['job_posted_date'] = pd.to_datetime(df_original['job_posted_date'])
df_altered = df_original.copy()
```

```python
df_altered.pivot_table(index='job_title_short', aggfunc='size')
```

```python
df_altered.groupby('job_title_short').size()
```

```python
df_altered.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc='median')
```

```python
df_altered.groupby('job_title_short')['salary_year_avg'].median()
```

```python
df_altered.pivot_table(values='salary_year_avg', index='job_title_short', columns='job_country', aggfunc='median')
```

```python
top_countries = df_altered['job_country'].value_counts().head(6).index
top_countries
df_job_country_salary = df_altered.pivot_table(
    values='salary_year_avg', 
    index='job_country', 
    columns='job_title_short', 
    aggfunc='median')
job_titles = ['Data Analyst', 'Data Engineer', 'Data Scientist']
df_job_country_salary = df_job_country_salary.loc[top_countries]
df_job_country_salary = df_job_country_salary[job_titles]
df_job_country_salary.plot(kind='bar')
plt.xticks(rotation=45, ha = 'right')
plt.title(label='Median Salary by Country and Job Title')
plt.ylabel('Median Salary (USD $)')
plt.xlabel('')
plt.show()

```
