```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df_original = dataset['train'].to_pandas()
df_altered = df_original.copy()

# Cleansing data

df_altered['job_posted_date'] = pd.to_datetime(df_altered['job_posted_date'])
df_altered
```

```python
df_US = df_altered[df_altered['job_country']=='United States']
df_US['job_posted_month'] = df_US['job_posted_date'].dt.strftime('%B')
df_US
```

```python
df_US.pivot_table(
    index='job_posted_month', 
    columns='job_title_short',
    aggfunc='size')
```

```python
df_US_pivot = df_US.pivot_table(
    index='job_posted_month', 
    columns='job_title_short',
    aggfunc='size'
    )

df_US_pivot.reset_index(inplace=True)
df_US_pivot['month_no'] = pd.to_datetime(df_US_pivot['job_posted_month'], format='%B').dt.month
df_US_pivot.sort_values('month_no', inplace=True) 
df_US_pivot.set_index('job_posted_month', inplace=True)
df_US_pivot.drop(columns='month_no', inplace=True)
top_3_jobs = df_US['job_title_short'].value_counts().head(3).index
df_US_pivot = df_US_pivot[top_3_jobs]
df_US_pivot.plot(kind='line')
# or top_3_jobs = df_US['job_title_short'].value_counts().head(3) top_3_jobs = top_3_jobs.index.tolist() df_US_pivot[top_3_kobs].plot
plt.ylabel('job Count')
plt.xlabel(2023)
plt.title('Monthly Job Postings for top Data Jobs in the US')
plt.legend()
plt.xticks(rotation=25)
plt.show()
```
